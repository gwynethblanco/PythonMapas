### gráficos descriptivo

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # Crear una figura con 1 fila y 3 columnas

sns.countplot(x="GRAVEDAD_ACCIDENTE", data=accidbaq, palette="twilight", ax=axes[0])
axes[0].set_title("Gravedad del Accidente")
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, fontsize=8)

sns.countplot(x="MES_ACCIDENTE", data=accidbaq,palette="twilight", ax=axes[1])
axes[1].set_title("Mes del Accidente")
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, fontsize=8)

sns.countplot(x="AÑO_ACCIDENTE", data=accidbaq, palette="twilight", ax=axes[2])
axes[2].set_title("Año del Accidente")
axes[2].set_xticklabels(axes[2].get_xticklabels(), rotation=45, fontsize=8)
 
plt.tight_layout()
plt.show()
```

Después de analizar los gráficos anteriores, podemos concluir que, en términos de gravedad de los accidentes, la categoría `Solo daños` tuvo la mayor cantidad de observaciones, con alrededor de 15,000, mientras que la categoría ``Con muertos`` tuvo la menor cantidad de observaciones, lo cual es algo positivo.
Por otro lado, el gráfico de mes de accidentes muestra que los tres primeros meses del año (enero, febrero y marzo) fueron los meses con más accidentes, con 2,200 a 2,500 observaciones.
Finalmente, en el gráfico del año del accidente, se puede ver que 2018 y 2019 fueron los años con las observaciones más altas, con aproximadamente 5,900 y 6,000, respectivamente, mientras que 2023 y el año en curso, 2024, presentaron las observaciones más bajas.

```python

fig, axs = plt.subplots(1, 3, figsize=(20, 6))  

sns.boxplot(y="AÑO_ACCIDENTE", data=accidbaq, color="#483D8B", ax=axs[0])
axs[0].set_title("Cantidad de Accidentes por Año")
axs[0].set_ylabel("Cantidad de Accidentes")

accidentes_agg = accidbaq['CANTIDAD_ACCIDENTES'].value_counts().reset_index()
accidentes_agg.columns = ['Cantidad Accidentes', 'Cantidad']
sns.barplot(x="Cantidad Accidentes", y="Cantidad", data=accidentes_agg, color="#483D8B", ax=axs[1])
axs[1].set_title("Cantidad de Accidentes")
axs[1].set_ylabel("Cantidad")
axs[1].set_xlabel("Cantidad Accidentes")

accidbaq_notna = accidbaq[accidbaq["CANT_HERIDOS_EN_SITIO_ACCIDENTE"].notna()]
sns.boxplot(y="CANT_HERIDOS_EN_SITIO_ACCIDENTE", data=accidbaq_notna, color="#483D8B", ax=axs[2])
axs[2].set_title("Cantidad de Heridos en el Sitio del Accidente")
axs[2].set_ylabel("Cantidad de Heridos")

plt.tight_layout()
plt.show()
```

Luego de efectuar los gráficos anteriores, podemos concluir, a partir del diagrama de cajas de la `Cantidad de accidentes por año`, que los datos presentan una ligera asimetría hacia la derecha, con la mediana en el año 2020. Además, no se observan datos atípicos, y la mayoría de los datos se concentran entre los años 2019 y 2021.
Por otro lado, en la segunda gráfica, que corresponde a un diagrama de barras que muestra la cantidad de accidentes, se observa que todos los datos se encuentran en 1, lo que sugiere que la mayoría de los sitios reportaron, como máximo, un accidente.
Por último, en el diagrama de cajas para la `Cantidad de heridos en el sitio del accidente`, se pueden notar algunos datos atípicos, lo que implica que en ciertos lugares se detectó un mayor número de heridos. Sin embargo, en la mayoría de los sitios, se presentó un número menor de heridos.

### Gráficas

Luego de efectuar el anterior gráfico, podemos observar que en departamenento en donde hubo más aumento del precio del combustible fue en el *Archipielago de San Andres*. Mientras que los demás estuvieron en un rango de precios muy parecidos entre si.
